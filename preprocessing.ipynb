## Tien xu li du lieu cac cot can thiet

```python
import pandas as pd

movies = pd.read_csv("./Data/Data_final/5232movies_imdb.csv", encoding='latin1')
movies
```

```python
movies.isnull().sum()
```

```python
movies= movies.drop_duplicates(subset=['Title'])
movies = movies.reset_index(drop=True)
```

### xu li cot year

```python
movies['Year'].unique()
```

```python
movies['Year'] = movies['Year'].str.extract(r'(\d{4})') 
movies['Year'] = movies['Year'].astype(int)

movies
```

### Xử lí cột Runtime

```python
movies['Runtime'] = movies['Runtime'].str.replace(' min', '')
movies['Runtime'] = movies['Runtime'].str.replace(',', '').astype(float).astype('Int64')
movies
```

### Xử lí cột Genre

```python
movies['Genre'] = movies['Genre'].fillna('')
genre_split = movies['Genre'].str.split(', ', expand=True)
genre_split.columns = ['Genre1', 'Genre2', 'Genre3']
movies = pd.concat([movies, genre_split], axis=1)
movies
```

### Xử lí cột Stars

```python
temp_df = movies['Stars'].str.split(', ', expand=True)

# Tạo các cột mới từ các cột tạm thời
movies['Stars1'] = temp_df[0]
movies['Stars2'] = temp_df[1]
movies['Stars3'] = temp_df[2]

movies
```

```python
movies_use = movies[['Title', 'Year', 'Runtime', 'Genre1', 'Genre2', 'Genre3', 'Stars1', 'Stars2', 'Stars3', 'Description', 'Rating', 'Votes', 'Img_link']]
movies_use
```

## Loc theo noi dung phim

```python
movies_test = movies[['Title','Year', 'Description']]
movies_test
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

movies_test = movies[['Title','Year', 'Description']]
movies_test


# Biểu diễn mô tả phim dưới dạng vectơ TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies_test["Description"].fillna(""))

# Tính toán độ tương đồng giữa các phim
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def recommend_movies(movie_title, top_n=5):
    # Kiểm tra xem tên phim có trong dữ liệu không
    if movie_title not in movies_test['Title'].values:
        return "Phim không tồn tại trong dữ liệu."
    
    # Tìm index của phim đầu vào
    idx = movies_test.index[movies_test['Title'] == movie_title][0]
    
    # Lấy điểm tương đồng của phim đầu vào với tất cả các phim khác
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sắp xếp các phim theo điểm tương đồng
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Lấy top N phim tương đồng nhất
    sim_scores = sim_scores[1:top_n+1]
    
    # Lấy thông tin của các phim tương đồng nhất
    similar_movies = []
    for i in sim_scores:
        title = movies_test.iloc[i[0]]['Title']
        year = movies_test.iloc[i[0]]['Year'] if 'Year' in movies_test.columns else "Năm sản xuất không xác định"
        similar_movies.append((title, year))
    
    return similar_movies

# Test hệ thống đề xuất
movie_title = "New Man Haha"  # Thay bằng tên phim bạn muốn kiểm tra
num_recommendations = 3  # Số lượng phim muốn gợi ý
recommended_movies = recommend_movies(movie_title, num_recommendations)
print(f"Các phim được đề xuất cho '{movie_title}':")
for title, year in recommended_movies:
    print(f"- {title} ({year})")
```

## Loc theo su tuong quan giua cac user voi nhau

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

movies_data = movies[['Title', 'Year', 'Rating', 'Director', 'Stars', 'Description']]

# Loại bỏ dấu câu từ các trường dữ liệu
movies_data['Description'] = movies_data['Description'].fillna("").str.lower()  # Chuyển đổi thành chữ thường
movies_data['Description'] = movies_data['Description'].str.replace('[^\w\s]', '')  # Loại bỏ dấu câu
movies_data['Description'] = movies_data['Description'].str.replace('\d+', '')  # Loại bỏ chữ số

# Kết hợp các thông tin của phim thành một cột text
movies_data['combined_features'] = movies_data['Title'] + " " + \
                                   movies_data['Director'] + " " + \
                                   movies_data['Stars'] + " " + \
                                   movies_data['Description']

# Thay thế giá trị NaN bằng chuỗi rỗng
movies_data['combined_features'].fillna("", inplace=True)

# Biểu diễn dữ liệu dưới dạng vectơ TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies_data['combined_features'])

# Tính toán độ tương đồng giữa các phim
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def recommend_movies(movie_title, top_n=5):
    # Tìm index của phim đầu vào
    idx = movies_data.index[movies_data['Title'] == movie_title][0]
    
    # Lấy điểm tương đồng của phim đầu vào với tất cả các phim khác
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sắp xếp các phim theo điểm tương đồng
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Lấy top N phim tương đồng nhất
    sim_scores = sim_scores[1:top_n+1]
    
    # Lấy thông tin của các phim tương đồng nhất
    similar_movies = [(movies_data.iloc[i[0]]['Title'], movies_data.iloc[i[0]]['Year'], movies_data.iloc[i[0]]['Rating']) for i in sim_scores]
    
    return similar_movies

# Test hệ thống đề xuất
movie_title = "New Man Haha"  # Thay bằng tên phim bạn muốn kiểm tra
num_recommendations = 3  # Số lượng phim muốn gợi ý
recommended_movies = recommend_movies(movie_title, num_recommendations)
print(f"Các phim được đề xuất cho '{movie_title}':")
for title, year, rating in recommended_movies:
    print(f"- {title} ({year}) - Điểm: {rating}")
```
