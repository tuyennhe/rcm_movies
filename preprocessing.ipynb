```python
import pandas as pd

movies = pd.read_csv("./Data/Data_final/5232movies_imdb.csv", encoding='latin1')
movies
```

```python
movies= movies.drop_duplicates(subset=['Title'])
movies = movies.reset_index(drop=True)
```

```python
movies['Year'] = movies['Year'].str.extract(r'(\d{4})') 
movies['Year'] = movies['Year'].astype(int)

movies
```

### Xử lí cột Runtime

```python
movies['Runtime'] = movies['Runtime'].str.replace(' min', '')
movies['Runtime'] = movies['Runtime'].str.replace(',', '').astype(float).astype('Int64')
movies
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Tải trình phân tích cảm xúc VADER
nltk.download('vader_lexicon')

# Khởi tạo SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Dùng VADER để xác định cảm xúc của mỗi mô tả phim
def analyze_sentiment(description):
    # Phân tích cảm xúc
    scores = sid.polarity_scores(description)
    # Lấy điểm tích cực, tiêu cực và trung lập
    positive_score = scores['pos']
    negative_score = scores['neg']
    # Đánh giá cảm xúc dựa trên điểm số
    if positive_score > negative_score:
        return "Tích cực"
    elif negative_score > positive_score:
        return "Tiêu cực"
    else:
        return "Trung lập"

# Một số mô tả phim trong dữ liệu của bạn
descriptions = [
    "A French-American in Paris lives by sponging off his working friends, and throws a party using borrowed money when his rich American aunt dies, believing firmly in his horoscope.",
    "The early career of cardiac surgeon Zbigniew Religa. Despite the harsh reality of 1980s Poland, he successfully leads a team of doctors to the country's first human heart transplantation.",
    "An homage to the work of psychologist Wilhelm Reich, matched with a story about a Yugoslavian girl's affair with a Russian skater. Sexual repression, social systems and the orgone theory are explored.",
    "Two rival test pilots strive to be the best there is at what they do and also to win the heart of the woman behind the AI pop sensation Sharon Apple. Movie compilation of the acclaimed four-chapter OVA with additional scenes.",
    "Cameramen and women discuss the craft and art of cinematography, illustrating their points with clips from 100 films, from The Birth of a Nation (1915) to Do the Right Thing (1989)."
]

# Phân tích cảm xúc cho mỗi mô tả phim
for description in descriptions:
    sentiment = analyze_sentiment(description)
    print(f"Mô tả: {description}")
    print(f"Cảm xúc: {sentiment}")
    print()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Tải trình phân tích cảm xúc VADER
nltk.download('vader_lexicon')

# Khởi tạo SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()


# Tạo một hàm để đánh nhãn cảm xúc của mỗi mô tả phim
def label_sentiment(description):
    # Phân tích cảm xúc
    scores = sid.polarity_scores(description)
    # Lấy điểm số compound
    compound_score = scores['compound']
    
    # Gán nhãn cảm xúc dựa trên điểm số compound
    if compound_score >= 0.05:
        sentiment = "Positive"
    elif compound_score <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    return sentiment

# Gán nhãn cảm xúc cho mỗi phim
movies['Sentiment'] = movies['Description'].apply(label_sentiment)

# Hiển thị DataFrame với cột Sentiment
print(movies[['Title', 'Sentiment']])

```

```python
positive_threshold = 7
neutral_threshold_low = 4
neutral_threshold_high = 7

# Hàm áp dụng quy tắc
def recommend_movie(rating, sentiment):
    if rating >= positive_threshold or (rating >= neutral_threshold_high and sentiment == "Neutral"):
        return "Recommend"
    else:
        return "Not Recommend"

# Áp dụng hàm recommend_movie cho mỗi hàng trong DataFrame
movies['Recommendation'] = movies.apply(lambda row: recommend_movie(row['Rating'], row['Sentiment']), axis=1)

# Hiển thị DataFrame với cột Recommendation
print(movies[['Title', 'Rating', 'Sentiment', 'Recommendation']])

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

movies_test = movies[['Title','Year', 'Description']]
movies_test


# Biểu diễn mô tả phim dưới dạng vectơ TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies_test["Description"].fillna(""))

# Tính toán độ tương đồng giữa các phim
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def recommend_movies(movie_title, top_n=5):
    # Kiểm tra xem tên phim có trong dữ liệu không
    if movie_title not in movies_test['Title'].values:
        return "Phim không tồn tại trong dữ liệu."
    
    # Tìm index của phim đầu vào
    idx = movies_test.index[movies_test['Title'] == movie_title][0]
    
    # Lấy điểm tương đồng của phim đầu vào với tất cả các phim khác
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sắp xếp các phim theo điểm tương đồng
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Lấy top N phim tương đồng nhất
    sim_scores = sim_scores[1:top_n+1]
    
    # Lấy thông tin của các phim tương đồng nhất
    similar_movies = []
    for i in sim_scores:
        title = movies_test.iloc[i[0]]['Title']
        year = movies_test.iloc[i[0]]['Year'] if 'Year' in movies_test.columns else "Năm sản xuất không xác định"
        similar_movies.append((title, year))
    
    return similar_movies

# Test hệ thống đề xuất
movie_title = "New Man Haha"  # Thay bằng tên phim bạn muốn kiểm tra
num_recommendations = 3  # Số lượng phim muốn gợi ý
recommended_movies = recommend_movies(movie_title, num_recommendations)
print(f"Các phim được đề xuất cho '{movie_title}':")
for title, year in recommended_movies:
    print(f"- {title} ({year})")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

movies_data = movies[['Title', 'Year', 'Rating', 'Director', 'Stars', 'Description']]

# Loại bỏ dấu câu từ các trường dữ liệu
movies_data['Description'] = movies_data['Description'].fillna("").str.lower()  # Chuyển đổi thành chữ thường
movies_data['Description'] = movies_data['Description'].str.replace('[^\w\s]', '')  # Loại bỏ dấu câu
movies_data['Description'] = movies_data['Description'].str.replace('\d+', '')  # Loại bỏ chữ số

# Kết hợp các thông tin của phim thành một cột text
movies_data['combined_features'] = movies_data['Title'] + " " + \
                                   movies_data['Director'] + " " + \
                                   movies_data['Stars'] + " " + \
                                   movies_data['Description']

# Thay thế giá trị NaN bằng chuỗi rỗng
movies_data['combined_features'].fillna("", inplace=True)

# Biểu diễn dữ liệu dưới dạng vectơ TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies_data['combined_features'])

# Tính toán độ tương đồng giữa các phim
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def recommend_movies(movie_title, top_n=5):
    # Tìm index của phim đầu vào
    idx = movies_data.index[movies_data['Title'] == movie_title][0]
    
    # Lấy điểm tương đồng của phim đầu vào với tất cả các phim khác
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sắp xếp các phim theo điểm tương đồng
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Lấy top N phim tương đồng nhất
    sim_scores = sim_scores[1:top_n+1]
    
    # Lấy thông tin của các phim tương đồng nhất
    similar_movies = [(movies_data.iloc[i[0]]['Title'], movies_data.iloc[i[0]]['Year'], movies_data.iloc[i[0]]['Rating']) for i in sim_scores]
    
    return similar_movies

# Test hệ thống đề xuất
movie_title = "New Man Haha"  # Thay bằng tên phim bạn muốn kiểm tra
num_recommendations = 3  # Số lượng phim muốn gợi ý
recommended_movies = recommend_movies(movie_title, num_recommendations)
print(f"Các phim được đề xuất cho '{movie_title}':")
for title, year, rating in recommended_movies:
    print(f"- {title} ({year}) - Điểm: {rating}")
```
